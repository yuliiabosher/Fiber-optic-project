Geopandas, Pandas and Matplotlib libraries were imported

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

The shape file with geometries of the UK parliamentary constituencies was downloaded and uploaded to Github and then read using the Github link. The shape file can be accessed at https://www.data.gov.uk/dataset/76a9c59b-647f-46cc-8511-1cb6d52e5f1d/westminster-parliamentary-constituencies-december-2021-boundaries-uk-buc, Data Links -> Shapefile.  The resulting geodataframe of the UK parliamentary constituencies is as follows


```python
constituencies = gpd.read_file('https://github.com/yuliiabosher/Fiber-optic-project/raw/refs/heads/parliamentary-constituencies/Westminster_Parliamentary_Constituencies_Dec_2021_UK_BUC_2022_-8882165546947265805.zip')
display(constituencies.head())
```

The columns 'PCON21CD', 'PCON21NM', 'geometry' were separated from the original geodataframe of the UK parliamentary constituencies to create a new cleaned geodataframe

```python
constituencies_cleaned = constituencies[['PCON21CD', 'PCON21NM', 'geometry']]
display(constituencies_cleaned.head())
```

The file on 2024 Fixed coverage parliamentary constituency data was downloaded from https://www.ofcom.org.uk/phones-and-broadband/coverage-and-speeds/connected-nations-update-spring-2024/, third row "Fixed coverage parliamentary constituency data". Then it was uploaded to Github and read using the Github link. The resulting dataframe is as follows

```python
ofcom = pd.read_csv("https://raw.githubusercontent.com/yuliiabosher/Fiber-optic-project/refs/heads/parliamentary-constituencies/202401_fixed_pcon_coverage_r01.csv", \
encoding="latin")
display(ofcom.head())
```

The columns in the original Ofcom dataframe are as follows

```python
display(ofcom.columns)
```

The columns 'parl_const',  'parliamentary_constituency_name',  'All Premises',   'Number of premises with Full Fibre availability' were separated from the original Ofcom dataframe to create a new cleaned dataframe

```python
ofcom_full_fibre = ofcom[['parl_const', 'parliamentary_constituency_name','All Premises', 'Number of premises with Full Fibre availability']]
display(ofcom_full_fibre.head())
```

Read in the urban/rural classification file

```python
import os.path
import errno

#mount google drive
from google.colab import drive
drive.mount('/content/drive')

#############
# Urban Rural Classification file at parliamentary constituency level available from https://pages.mysociety.org/uk_ruc/datasets/uk_ruc/latest
#############
fileUrbanRuralClassification = '/content/drive/MyDrive/Colab/pcon_ruc.csv'

if os.path.exists(fileUrbanRuralClassification) :
    print("Reading Urban/Rural classification.....")
    dfRUC = pd.read_csv(fileUrbanRuralClassification)
else:
    raise FileNotFoundError(errno.ENOENT, os.strerror(errno.ENOENT), fileUrbanRuralClassification)
print("Urban Rural Classification file data shape:", dfRUC.shape)
```

Show the columns and select which ones we want

```python
display(dfRUC.columns)

dfRUC = dfRUC[['gss-code', 'ruc-cluster-label']]
display(dfRUC.head())
```

The number of rows, i.e. the number of parliamentary constituencies in the two cleaned datasets are the same, which means they could be joined without any discrepancies

```python
display(f'Number of rows in the geodataframe of the UK constituencies - {len(constituencies_cleaned)}')
display(f'Number of rows in the Ofcom dataframe on Fill Fibre availability by constituency - {len(ofcom_full_fibre)}')
display(f'Number of rows in the Urban Rural dataframe by constituency - {len(dfRUC)}')
```

The geodatframe of the UK parliamentary constituencies and the dataframe on full fibre by parliamentary constituency were joined on parliamentary constituency code column

```python
fibre_by_constituency_geo = constituencies_cleaned.merge(ofcom_full_fibre, left_on='PCON21CD', right_on='parl_const')
fibre_by_constituency_geo = fibre_by_constituency_geo.merge(dfRUC, left_on='PCON21CD', right_on='gss-code')
display(fibre_by_constituency_geo.head())


```

The joined dataframe had the same number of rows as the original two, which means no constituency was left out

```python
display(f'The legth of the merged dataframe is {len(fibre_by_constituency_geo)}')
```

The columns with duplicated data in the joined geodataframe were dropped

```python
fibre_by_constituency_geo.drop('parl_const', axis=1, inplace=True)
fibre_by_constituency_geo.drop('parliamentary_constituency_name', axis=1, inplace=True)
fibre_by_constituency_geo.drop('gss-code', axis=1, inplace=True)
display(fibre_by_constituency_geo.head())
```

The columns in the joined geodataframe were renamed

```python
fibre_by_constituency_geo.columns = ['Constituency Code', 'Constituency Name', 'geometry', 'Total Premises', 'Premises with Full Fibre Availability', 'Urban/Rural Classification']
display(fibre_by_constituency_geo.head())
```

It was verified that 'Total Premises' and 'Premises with Full Fibre Availability' columns were integer data type

```python
display(fibre_by_constituency_geo.info())
```

A new column titled 'Percentage of Premises with Full Fibre Availability' was added. The column was the result of 'Premises with Full Fibre Availability' column divided by 'Total Premises' column



```python
fibre_by_constituency_geo['Percentage of Premises with Full Fibre Availability'] = \
fibre_by_constituency_geo['Premises with Full Fibre Availability'] / fibre_by_constituency_geo['Total Premises']
display(fibre_by_constituency_geo.head())
```

The 'Percentage of Premises with Full Fibre Availability' column was plotted as a cholopleth

```python
fig, ax = plt.subplots(1, 2, figsize=(24,24))
fibre_by_constituency_geo.plot(column='Percentage of Premises with Full Fibre Availability', cmap='Blues', legend=True, ax=ax[0], legend_kwds={'shrink': 0.7})
fig.set_size_inches(16,12)
plt.style.use('seaborn-v0_8-whitegrid')
font = {'family':'monospace','color':'black','size':12}
ax[0].set_title('Percentage of Premises with \nFull Fibre Availability by Constituency', fontdict = font)


fibre_by_constituency_geo.plot(column='Urban/Rural Classification', cmap='Set2', legend=True, ax=ax[1], legend_kwds={'frameon': True, 'loc': 'upper left'})
ax[1].set_title('Urban/Rural Classification of Parliamentary Constituencies', fontdict = font)
plt.axis('off')
plt.show()
```
