Imported Geopandas, Pandas and Matplotlib Libraries

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

The function to read a geojson file was defined

```python
def read_geojson_file(file):
  df = gpd.read_file(file)
  return df
```

The link to the geojson file with geometries of the UK parliamentary constituencies can be accessed at https://www.data.gov.uk/dataset/76a9c59b-647f-46cc-8511-1cb6d52e5f1d/westminster-parliamentary-constituencies-december-2021-boundaries-uk-buc. The original geodataframe of the UK parliamentary constituencies is as follows


```python
constituencies = read_geojson_file('https://open-geography-portalx-ons.hub.arcgis.com/api/download/v1/items/a74f3451133140e2a49609f95db2ea64/geojson?layers=0')
display(constituencies.head())
```

The columns 'PCON21CD', 'PCON21NM', 'geometry' were separated from the original geodataframe of the UK parliamentary constituencies to create a new cleaned geodataframe as follows

```python
constituencies_cleaned = constituencies[['PCON21CD', 'PCON21NM', 'geometry']]
display(constituencies_cleaned.head())
```

The link to the file can be accessed at https://www.ofcom.org.uk/phones-and-broadband/coverage-and-speeds/data/, Fixed coverage parliamentary constituency data (ZIP, 100.6 KB). The original Ofcom dataframe is as follows

```python
ofcom = pd.read_csv("https://raw.githubusercontent.com/yuliiabosher/Fiber-optic-project/refs/heads/parliamentary-constituencies/202209_fixed_pcon_coverage_r02.csv",encoding="latin")
display(ofcom.head())
```

The columns in the original Ofcom dataframe are as follows

```python
display(ofcom.columns)
```

The columns 'parl_const',  'parliamentary_constituency_name',  'All Premises',   'Number of premises with Full Fibre availability' were separated from the original Ofcom dataframe to create a new cleaned dataframe

```python
ofcom_full_fibre = ofcom[['parl_const', 'parliamentary_constituency_name','All Premises', 'Number of premises with Full Fibre availability']]
display(ofcom_full_fibre.head())
```

The number of rows, i.e. the number of parliamentary constituencies in the two cleaned datasets are the same, which means we can join them without any discrepancies

```python
display(f'Number of rows in the geodataframe of the UK constituencies - {len(constituencies_cleaned)}')
display(f'Number of rows in the Ofcom dataframe on Fill Fibre availability by constituency - {len(ofcom_full_fibre)}')
```

The geodatframe of the UK parliamentary constituencies and the dataframe on full fibre by parliamentary constituency were joined on parliamentary constituency code column

```python
fibre_by_constituency_geo = constituencies_cleaned.merge(ofcom_full_fibre, left_on='PCON21CD', right_on='parl_const')
display(fibre_by_constituency_geo.head())
```

The joined dataframe had the same number of rows as the original two, which means no constituency was left out

```python
display(f'The legth of the merged dataframe is {len(fibre_by_constituency_geo)}')
```

Dropped the columns with duplicated data in the joined dataframe

```python
fibre_by_constituency_geo.drop('parl_const', axis=1, inplace=True)
fibre_by_constituency_geo.drop('parliamentary_constituency_name', axis=1, inplace=True)
display(fibre_by_constituency_geo.head())
```

Renamed the columns in the joined dataframe

```python
fibre_by_constituency_geo.columns = ['Constituency Code', 'Constituency Name', 'geometry', 'Total Premises', 'Premises with Full Fibre Availability']
display(fibre_by_constituency_geo.head())
```

Verified that 'Total Premises' and 'Premises with Full Fibre Availability' are integer type

```python
display(fibre_by_constituency_geo.info())
```

Added a new column titled 'Percentage of Premises with Full Fibre Availability' which is the result of 'Premises with Full Fibre Availability' column divided by 'Total Premises' column



```python
fibre_by_constituency_geo['Percentage of Premises with Full Fibre Availability'] = \
fibre_by_constituency_geo['Premises with Full Fibre Availability'] / fibre_by_constituency_geo['Total Premises']
display(fibre_by_constituency_geo.head())
```

Plotted the 'Percentage of Premises with Full Fibre Availability' column on a heatmap

```python
fig, ax = plt.subplots(1, 1)
fibre_by_constituency_geo.plot(column='Percentage of Premises with Full Fibre Availability', cmap='Blues', legend=True, ax=ax, legend_kwds={'shrink': 0.7})
fig.set_size_inches(8,6)
plt.style.use('seaborn-v0_8-whitegrid')
font = {'family':'monospace','color':'black','size':12}
ax.set_title('Percentage of Premises with \nFull Fibre Availability by Constituency', fontdict = font)
plt.axis('off')
plt.show()
```
