Geopandas, Pandas and Matplotlib libraries were imported

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

The shape file with geometries of the UK parliamentary constituencies was downloaded and uploaded to Github and then read using the Github link. The shape file can be accessed at https://www.data.gov.uk/dataset/76a9c59b-647f-46cc-8511-1cb6d52e5f1d/westminster-parliamentary-constituencies-december-2021-boundaries-uk-buc, Data Links -> Shapefile.  The resulting geodataframe of the UK parliamentary constituencies is as follows


```python
constituencies = gpd.read_file('https://github.com/yuliiabosher/Fiber-optic-project/raw/refs/heads/parliamentary-constituencies/Westminster_Parliamentary_Constituencies_Dec_2021_UK_BUC_2022_-8882165546947265805.zip')
display(constituencies.head())
```

The columns 'PCON21CD', 'PCON21NM', 'geometry' were separated from the original geodataframe of the UK parliamentary constituencies to create a new cleaned geodataframe

```python
constituencies_cleaned = constituencies[['PCON21CD', 'PCON21NM', 'geometry']]
display(constituencies_cleaned.head())
```

The file on 2024 Fixed coverage parliamentary constituency data was downloaded from https://www.ofcom.org.uk/phones-and-broadband/coverage-and-speeds/connected-nations-update-spring-2024/, third row "Fixed coverage parliamentary constituency data". Then it was uploaded to Github and read using the Github link. The resulting dataframe is as follows

```python
ofcom = pd.read_csv("https://raw.githubusercontent.com/yuliiabosher/Fiber-optic-project/refs/heads/parliamentary-constituencies/202401_fixed_pcon_coverage_r01.csv", \
encoding="latin")
display(ofcom.head())
```

The columns in the original Ofcom dataframe are as follows

```python
display(ofcom.columns)
```

The columns 'parl_const',  'parliamentary_constituency_name',  'All Premises',   'Number of premises with Full Fibre availability' were separated from the original Ofcom dataframe to create a new cleaned dataframe

```python
ofcom_full_fibre = ofcom[['parl_const', 'parliamentary_constituency_name','All Premises', 'Number of premises with Full Fibre availability']]
display(ofcom_full_fibre.head())
```

The number of rows, i.e. the number of parliamentary constituencies in the two cleaned datasets are the same, which means they could be joined without any discrepancies

```python
display(f'Number of rows in the geodataframe of the UK constituencies - {len(constituencies_cleaned)}')
display(f'Number of rows in the Ofcom dataframe on Fill Fibre availability by constituency - {len(ofcom_full_fibre)}')
```

The geodatframe of the UK parliamentary constituencies and the dataframe on full fibre by parliamentary constituency were joined on parliamentary constituency code column

```python
fibre_by_constituency_geo = constituencies_cleaned.merge(ofcom_full_fibre, left_on='PCON21CD', right_on='parl_const')
display(fibre_by_constituency_geo.head())
```

The joined dataframe had the same number of rows as the original two, which means no constituency was left out

```python
display(f'The legth of the merged dataframe is {len(fibre_by_constituency_geo)}')
```

The columns with duplicated data in the joined geodataframe were dropped

```python
fibre_by_constituency_geo.drop('parl_const', axis=1, inplace=True)
fibre_by_constituency_geo.drop('parliamentary_constituency_name', axis=1, inplace=True)
display(fibre_by_constituency_geo.head())
```

The columns in the joined geodataframe were renamed

```python
fibre_by_constituency_geo.columns = ['Constituency Code', 'Constituency Name', 'geometry', 'Total Premises', 'Premises with Full Fibre Availability']
display(fibre_by_constituency_geo.head())
```

It was verified that 'Total Premises' and 'Premises with Full Fibre Availability' columns were integer data type

```python
display(fibre_by_constituency_geo.info())
```

A new column titled 'Percentage of Premises with Full Fibre Availability' was added. The column was the result of 'Premises with Full Fibre Availability' column divided by 'Total Premises' column



```python
fibre_by_constituency_geo['Percentage of Premises with Full Fibre Availability'] = \
fibre_by_constituency_geo['Premises with Full Fibre Availability'] / fibre_by_constituency_geo['Total Premises']
display(fibre_by_constituency_geo.head())
```

The 'Percentage of Premises with Full Fibre Availability' column was plotted as a cholopleth

```python
fig, ax = plt.subplots(1, 1)
fibre_by_constituency_geo.plot(column='Percentage of Premises with Full Fibre Availability', cmap='Blues', legend=True, ax=ax, legend_kwds={'shrink': 0.7})
fig.set_size_inches(8,6)
plt.style.use('seaborn-v0_8-whitegrid')
font = {'family':'monospace','color':'black','size':12}
ax.set_title('Percentage of Premises with \nFull Fibre Availability by Constituency', fontdict = font)
plt.axis('off')
plt.show()
```

The file with percentages of highly rural, rural and urban areas in each parliamentary constituency can be accessed at https://pages.mysociety.org/uk_ruc/downloads/uk_ruc_xlsx/latest#survey

```python
uk_ruc = pd.read_excel('https://pages.mysociety.org/uk_ruc/data/uk_ruc/latest/uk_ruc.xlsx', sheet_name='pcon_ruc')
display(uk_ruc)
```

The dataframe with rural/urban data was merged with the dataframe on full fibre availability by parliamentary constituency.

```python
fibre_by_constituency_geo = fibre_by_constituency_geo.merge(uk_ruc, left_on='Constituency Code', right_on='gss-code')
display(fibre_by_constituency_geo.info())
```

Duplicate columns 'gss-code' and 'constituency-name' were dropped from the merged dataframe.

```python
fibre_by_constituency_geo.drop('gss-code', axis=1, inplace=True)
fibre_by_constituency_geo.drop('constituency-name', axis=1, inplace=True)
display(fibre_by_constituency_geo.head())
```

The newly merged columns were renamed.

```python
fibre_by_constituency_geo.columns = ['Constituency Code', 'Constituency Name', 'geometry', \
                                     'Total Premises', 'Premises with Full Fibre Availability', \
                                     'Percentage of Premises with Full Fibre Availability', \
                                     'Percentage of Highly Rural Areas', 'Percentage of Rural Areas', 'Percentage of Urban Areas', 'Classification Label']
display(fibre_by_constituency_geo.head())
```

Percentages of highly rural, rural and urban areas by constituency were plotted alongside full fibre availability to draw correlation.

```python
fig, ax = plt.subplots(1, 4)
fibre_by_constituency_geo.plot(column='Percentage of Premises with Full Fibre Availability', cmap='Blues', legend=True, ax=ax[0], legend_kwds={'shrink': 0.3})
fig.set_size_inches(20,10)
plt.style.use('seaborn-v0_8-whitegrid')
font = {'family':'monospace','color':'black','size':10}
ax[0].set_title('Percentage of Premises with \nFull Fibre Availability by Constituency', fontdict=font)
ax[0].axis('off')
ax[0].grid(visible=None)
fibre_by_constituency_geo.plot(column='Percentage of Highly Rural Areas', cmap='Greens', legend=True, ax=ax[1], legend_kwds={'shrink': 0.3})
ax[1].set_title('Percentage of Highly Rural areas \n by Constituency', fontdict=font)
ax[1].axis('off')
ax[1].grid(visible=None)
fibre_by_constituency_geo.plot(column='Percentage of Rural Areas', cmap='Oranges', legend=True, ax=ax[2], legend_kwds={'shrink': 0.3})
ax[2].set_title('Percentage of Rural areas \n by Constituency', fontdict=font)
ax[2].axis('off')
ax[2].grid(visible=None)
fibre_by_constituency_geo.plot(column='Percentage of Urban Areas', cmap='Purples', legend=True, ax=ax[3], legend_kwds={'shrink': 0.3})
ax[3].set_title('Percentage of Urban areas \n by Constituency', fontdict=font)
ax[3].axis('off')
ax[3].grid(visible=None)
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.hist(fibre_by_constituency_geo[fibre_by_constituency_geo['Classification Label']=='Urban']['Percentage of Premises with Full Fibre Availability'], color='purple', histtype='step')
plt.hist(fibre_by_constituency_geo[fibre_by_constituency_geo['Classification Label']=='Rural']['Percentage of Premises with Full Fibre Availability'], color='green', histtype='step')
plt.hist(fibre_by_constituency_geo[fibre_by_constituency_geo['Classification Label']=='Urban with rural areas']['Percentage of Premises with Full Fibre Availability'], color='grey', histtype='step')
plt.hist(fibre_by_constituency_geo[fibre_by_constituency_geo['Classification Label']=='Sparse and rural']['Percentage of Premises with Full Fibre Availability'], color='brown', histtype='step')
plt.legend(labels=['Urban', 'Rural', 'Urban with rural areas', 'Sparse and rural'])
plt.title('Percentage of Premises with Full Fibre Availability \nper Parliamentary Constituency by Classification Label', fontdict={'family':'monospace','color':'black','size':12})
plt.ylabel('Number of Parliamentary Constituencies', fontdict=font)
plt.xlabel('Percentage of Premises with Full Fibre Availability \nin a Parliamentary Constituency', fontdict=font)
plt.show()
```

```python
g = sns.regplot(data=fibre_by_constituency_geo, x='Percentage of Premises with Full Fibre Availability', y='Percentage of Urban Areas')
font = {'family':'monospace','color':'black','size':10}
plt.xticks(fontname="monospace")
plt.yticks(fontname="monospace")
plt.title('Percentage of Premises with Full Fibre Availability vs \nPercentage of Urban Areas by Parliamentary Constituency', fontdict = font)
plt.xlabel('Percentage of Premises with Full Fibre Availability')
plt.ylabel('Percentage of Urban Areas', fontdict = font)
```
